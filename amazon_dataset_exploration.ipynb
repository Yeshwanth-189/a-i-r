```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
from IPython.display import display
```

```python
df=pd.read_csv("amazon.csv")
```

```python
print("Dataset dimensions:", df.shape)
print(df.info())
```

```python
display(df.head())
```

```python
print("Missing values in each column:")
missing_values = df.isnull().sum()
print(missing_values)

# Optionally, see how many total missing values
total_missing = missing_values.sum()
print(f"\nTotal missing values in the dataset: {total_missing}")
```

```python
df['rating_count'] = df['rating_count'].str.replace(',', '')
df['rating_count'] = pd.to_numeric(df['rating_count'], errors='coerce')

median_rating_count = df['rating_count'].median()
df['rating_count'].fillna(median_rating_count, inplace=True)

print("Missing values in 'rating_count':", df['rating_count'].isnull().sum())
print("Data type:", df['rating_count'].dtype)
```

```python
df.drop_duplicates(inplace=True)

print("Shape after removing duplicate rows:", df.shape)
print("Unique product IDs:", df['product_id'].nunique())
print("Total rows:", len(df))

#Drop duplicate products based on product_id
df.drop_duplicates(subset='product_id', keep='first', inplace=True)
print("Shape after dropping duplicate product_ids:", df.shape)

df['user_id'] = df['user_id'].str.split(',').str[0]
df.to_csv("amazon.csv", index=False)
```

```python
df['rating'] = pd.to_numeric(df['rating'], errors='coerce')
# Remove ₹ and commas
df['discounted_price'] = df['discounted_price'].astype(str).str.replace('[₹$,]', '', regex=True)
df['actual_price'] = df['actual_price'].astype(str).str.replace('[₹$,]', '', regex=True)

# Convert to float
df['discounted_price'] = pd.to_numeric(df['discounted_price'], errors='coerce')
df['actual_price'] = pd.to_numeric(df['actual_price'], errors='coerce')
df['rating_count'] = df['rating_count'].astype(int)
display(df.dtypes)
```

```python
df['product_name'] = df['product_name'].astype(str).str.lower().str.strip()
```

```python
import re

def clean_text(text):
    if pd.isnull(text):
        return ''
    text = str(text).lower()                     # lowercase
    text = re.sub(r'<.*?>', '', text)            # remove HTML tags
    text = re.sub(r'[^\w\s]', '', text)          # remove punctuation
    text = re.sub(r'\s+', ' ', text).strip()     # remove extra spaces
    return text

df['about_product'] = df['about_product'].apply(clean_text)
df['review_content'] = df['review_content'].apply(clean_text)
display(df[['product_name', 'about_product', 'review_content']].head(3))
```

```python
df['price_difference'] = df['actual_price'] - df['discounted_price']
```

```python
df['is_discounted'] = (df['price_difference'] > 0).astype(int)
```

**Top Rated Mega Deals**

```python
top_discounted = df[(df['is_discounted'] == 1) & (df['rating'] >= 4)]
top_discounted = top_discounted.sort_values(by='price_difference', ascending=False)
display(top_discounted[['product_id','product_name', 'actual_price', 'discounted_price', 'price_difference','discount_percentage','rating','rating_count','about_product']].head(10))
top_discounted[
    ['product_name', 'actual_price', 'discounted_price', 'price_difference',
     'discount_percentage', 'rating', 'rating_count', 'about_product']
].head(10).to_csv("top_discounted_products.csv", index=False)
```

**Hot & Trending**

```python
# Step 1: Calculate popularity score (if not done already)
df['popularity_score'] = df['rating'] * df['rating_count']

# Step 2: Sort by category and popularity
popular_df = df.sort_values(
    by=['category', 'popularity_score'],
    ascending=[True, False]
)

popular_df=df[(df['rating'] >= 4.2) & (df['popularity_score'] > 2500)]

# Step 3: Pick top 1 (or N) product per category
top_popular_per_category = popular_df.groupby('category').head(1).reset_index(drop=True)

# Display popular picks
top_popular_per_category = (
    df.sort_values(by=['category', 'popularity_score'], ascending=[True, False])
    .groupby('category')
    .first()  # Picks the top product in each category
    .reset_index()
)
step_popular = top_popular_per_category.iloc[::20].reset_index(drop=True)
step_popular[
    ['product_name', 'product_id', 'rating', 'rating_count', 'popularity_score']
].head(10).to_csv("top_discounted_products.csv", index=False)
display(step_popular[['category', 'product_name', 'product_id', 'rating', 'rating_count', 'popularity_score']])
```

```python
!pip install vaderSentiment

from vaderSentiment.vaderSentiment import SentimentIntensityAnalyzer

analyzer = SentimentIntensityAnalyzer()

# Function to compute sentiment score
def get_sentiment(text):
    if not isinstance(text, str) or text.strip() == '':
        return 0
    score = analyzer.polarity_scores(text)['compound']
    return score  # Ranges from -1 (neg) to +1 (pos)

df['sentiment_score'] = df['review_content'].apply(get_sentiment)
```

**Customers love these**

```python
# Step 1: Filter products with good sentiment
high_sentiment_df = df[(df['rating'] >= 4) & (df['sentiment_score'] > 0.9)]

# Step 2: Sort by category, sentiment, rating, and rating_count
high_sentiment_df = high_sentiment_df.sort_values(
    by=['category', 'sentiment_score', 'rating', 'rating_count'],
    ascending=[True, False, False, False]
)

# Step 3: Pick top product per category
top_per_category = high_sentiment_df.groupby('category').head(1).reset_index(drop=True)

# Step 4: Select and export columns (include product_id this time!)
top_reviewed = top_per_category[
    ['category', 'product_name', 'product_id', 'rating', 'rating_count', 'sentiment_score']
]

# Step 5: Export to CSV (top 10 only if you like)
top_reviewed.head(10).to_csv("top_reviewed_products.csv", index=False)
display(top_reviewed.head(10))
```

**Hidden Gems**

```python
# Step 1: Filter products with strong sentiment score
high_sentiment_df = df[df['sentiment_score'] > 0.9]

# Step 2: Define low popularity threshold
low_rating_threshold = df['rating_count'].median()

# Step 3: Select high-sentiment & low-rated-count products
hidden_gems = high_sentiment_df[high_sentiment_df['rating_count'] < low_rating_threshold]

# Step 4: Sort by sentiment score
hidden_gems = hidden_gems.sort_values(by='sentiment_score', ascending=False)

# Step 5: Select relevant columns
hidden_gems = hidden_gems[
    ['product_name', 'product_id', 'category', 'rating', 'rating_count', 'sentiment_score']
]

# Step 6: Export top 10 to CSV
hidden_gems.head(10).to_csv("hidden_gems.csv", index=False)

# Optional: Display the result
display(hidden_gems.head(10))
```

**User Based Using KMeans**

```python
from surprise import Dataset, Reader, KNNBasic
reader = Reader(rating_scale=(1, 5))
data = Dataset.load_from_df(df[['user_id', 'product_id', 'rating']], reader)
import joblib
# Train/test split
trainset, testset = train_test_split(data, test_size=0.2)
```

```python
sim_options = {
    'name': 'cosine',
    'user_based': True   # Set to False for item-based
}

model = KNNBasic(sim_options=sim_options)
model.fit(trainset)
joblib.dump(model, 'user_knn_model.pkl')
```

```python
# Choose a user
user_id = 'AG3D6O4STAQKAY2UVGEUV46KN35Q'  # Replace with an actual user_id from your data

# Step 1: Get categories the user has interacted with
user_categories = df[df['user_id'] == user_id]['category'].value_counts().index.tolist()
# Step 2: Unseen products
# All products in dataset
all_products = df['product_id'].unique()
# Products already rated by this user
rated_products = df[df['user_id'] == user_id]['product_id'].unique()
unseen_products = [pid for pid in all_products if pid not in rated_products]
unseen_df = df[df['product_id'].isin(unseen_products)]

# Step 3: Keep only products from user's top categories
filtered_unseen_df = unseen_df[unseen_df['category'].isin(user_categories)].drop_duplicates('product_id')
# Step 4: Predict on unseen but same-category products
predictions = [
    model.predict(user_id, pid) for pid in filtered_unseen_df['product_id'].tolist()
]

# Step 5: Sort and get top N
top_predictions = sorted(predictions, key=lambda x: x.est, reverse=True)
top_product_ids = [pred.iid for pred in top_predictions[:10]]

# Step 6: Show the results
recommended_products = (
    df[df['product_id'].isin(top_product_ids)]
    .drop_duplicates('product_id')
    .sort_values(by='rating', ascending=False)
)

display(recommended_products[['product_name', 'category', 'rating']])


```
